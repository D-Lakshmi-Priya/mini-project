```python
import pandas as pd
import random
age=[]
height=[]
weight=[]
speed=[]
for i in range(100):
  age.append(random.randint(0,100))
  height.append(random.randint(0,100))
  weight.append(random.randint(0,100))
  speed.append(random.randint(0,100))

idata={'Age':age,'Height':height,'Weight':weight,'Speed':speed}

idata
```

```python
df1=pd.DataFrame(idata)
df1
```
